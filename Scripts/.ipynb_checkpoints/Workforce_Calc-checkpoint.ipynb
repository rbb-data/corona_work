```python
import pandas as pd
import numpy as np
import altair as alt
```

```python
data=pd.read_csv("../RawData/Physical_Proximity.csv")
data
```

```python
data["main_groups_code"]=data["Code"].str[:5]
listofmaingroups=list(data["main_groups_code"].unique())
print(len(listofmaingroups))
```

```python
contactwith_others=data.sort_values(by="main_groups_code")
contactwith_others
```

```python
groudict={}
index=0
for i in listofmaingroups:
    entrieslist=[]
    joblist=[(list(contactwith_others[contactwith_others["main_groups_code"]==i]["Occupation"])),i]
    metriclist=((contactwith_others[contactwith_others["main_groups_code"]==i]["Context"].describe().tolist()))
    entrieslist.extend(joblist)
    entrieslist.extend(metriclist)
    print(entrieslist)
    groudict[index]=entrieslist
    index+=1
groudict
```

```python
columnlist=["jobs","code"]
columnlist.extend((data["Context"].describe()).index.tolist())
columnlist
```

```python
dataframe=pd.DataFrame().from_dict(groudict).transpose()
dataframe
```

```python
dataframe.columns=columnlist
```

```python
dataframe=dataframe.fillna(0)
dataframe["rel_std"]=dataframe["std"]/dataframe["mean"]
dataframe.sort_values(by="rel_std")
```

```python
dataframe["rel_std"].mean()
```

```python
dataframe.to_csv("jobs_with_context.csv", index=False)
```
